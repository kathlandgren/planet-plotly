```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
hot_Jupiter_snapshot = pd.read_csv('./data/hot_Jupiter_snapshot.csv')
uniform_longitudes = pd.read_csv('./data/uniform_longitudes.csv')
gaussian_latitudes = pd.read_csv('./data/gaussian_latitudes.csv')
```

```python
theta = uniform_longitudes.values + np.pi
phi = np.arcsin(gaussian_latitudes.values)+np.pi/2
print(np.min(theta),np.max(theta))
print(np.min(phi),np.max(phi))
```

### Plotting temperature data on a sphere

```python
# this is a sphere
theta = uniform_longitudes.values + np.pi
phi = np.arcsin(gaussian_latitudes.values)+np.pi/2
#theta = np.append(theta,np.nan)
#phi = np.append(phi,np.nan)
x = np.outer(np.cos(theta),np.sin(phi))
y = np.outer(np.sin(theta),np.sin(phi))
z = np.outer(np.ones(127),np.cos(phi))

#call plotly
fig= go.Figure(go.Surface(x=x, y=y, z=z,surfacecolor=hot_Jupiter_snapshot.values[:,:-1].T,cmax=1200,cmin=700))
fig.show()
```
